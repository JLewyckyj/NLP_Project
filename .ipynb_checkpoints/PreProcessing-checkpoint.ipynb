```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import string
import numpy as np
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
#Read in data
data = pd.read_csv('Board_summary.csv')
print(data.shape)
data.head()
```

```python
#Pre-Processing for Text

#Convert Report to string
data['Report'] = data['Report'].astype(str)
#To lower case
data['Report'] = [x.lower() for x in data['Report']]
#Remove punctuation
data['Report'] = [''.join(c for c in s if c not in string.punctuation) for s in data['Report']]
#Remove numbers
pattern = '[0-9]'
data['Report'] = [re.sub(pattern, '', i) for i in data['Report']]
```

```python
#Check out distribution of labels

grouped_data = data.groupby(['FV'], as_index=False)['Report'].count()
print(grouped_data)
grouped_data.plot.bar(x='FV', y='Report', rot=0)
plt.title('Distribution of Labels')
plt.xlabel('Future Value Grade')
plt.ylabel('Count')
plt.savefig('label_distribution.png')
```

```python
#Recode Labels

##Convert any grades 50 or above (projected to be at least an average regular) in top group
data.loc[data['FV'] == '70', 'FV_r'] = 3
data.loc[data['FV'] == '65', 'FV_r'] = 3
data.loc[data['FV'] == '60', 'FV_r'] = 3
data.loc[data['FV'] == '55', 'FV_r'] = 3
data.loc[data['FV'] == '50', 'FV_r'] = 3

##Convert any grades 40+ to 45+ (projected to be MLB role player) in middle group
data.loc[data['FV'] == '45+', 'FV_r'] = 2
data.loc[data['FV'] == '45', 'FV_r'] = 2
data.loc[data['FV'] == '40+', 'FV_r'] = 2

##Convert any grades 40 or below (projected to be a very limited MLB role player or non-Major leaguer) in bottom group
data.loc[data['FV'] == '40', 'FV_r'] = 1
data.loc[data['FV'] == '35+', 'FV_r'] = 1

#New distribution of labels

grouped_data = data.groupby(['FV_r'], as_index=False)['Report'].count()
print(grouped_data)
grouped_data.plot.bar(x='FV_r', y='Report', rot=0)
plt.title('New Distribution of Labels')
plt.xlabel('Future Value Category')
plt.ylabel('Count')
plt.savefig('new_label_distribution.png')
```

```python
#Separate into pitchers and hitters
pit = data[(data['Pos'] == 'LHP') | (data['Pos'] == 'RHP')]
hit = data[(data['Pos'] != 'LHP') & (data['Pos'] != 'RHP')]
```

```python
#Tokenize the text

##Pitcher tokens
pit_tokens = []
for i in pit['Report']:
    word_list = word_tokenize(i)
    pit_tokens.append(word_list)
    
##Hitter tokens
hit_tokens = []
for i in hit['Report']:
    word_list = word_tokenize(i)
    hit_tokens.append(word_list)
```

```python
#tfidf Vectorizers

##Pitcher tfidf

####1gram
pit_tfidf1 = TfidfVectorizer(sublinear_tf=True, norm='l2', encoding='UTF-8', max_features=1000, ngram_range=(1,1), stop_words='english')
pit_tfidf1_features = pit_tfidf1.fit_transform(pit.Report).toarray()
print(pit_tfidf1_features.shape)
pit_tfidf1_features = pd.DataFrame(pit_tfidf1_features)

####1gram+2gram
pit_tfidf2 = TfidfVectorizer(sublinear_tf=True, norm='l2', encoding='UTF-8', max_features=1000, ngram_range=(1,2), stop_words='english')
pit_tfidf2_features = pit_tfidf2.fit_transform(pit.Report).toarray()
print(pit_tfidf2_features.shape)
pit_tfidf2_features = pd.DataFrame(pit_tfidf2_features)

####1gram+2gram+3gram
pit_tfidf3 = TfidfVectorizer(sublinear_tf=True, norm='l2', encoding='UTF-8', max_features=1000, ngram_range=(1,3), stop_words='english')
pit_tfidf3_features = pit_tfidf3.fit_transform(pit.Report).toarray()
print(pit_tfidf3_features.shape)
pit_tfidf3_features = pd.DataFrame(pit_tfidf3_features)

##Hitter tfidf

####1gram
hit_tfidf1 = TfidfVectorizer(sublinear_tf=True, norm='l2', encoding='UTF-8', max_features=1000, ngram_range=(1,1), stop_words='english')
hit_tfidf1_features = hit_tfidf1.fit_transform(hit.Report).toarray()
print(hit_tfidf1_features.shape)
hit_tfidf1_features = pd.DataFrame(hit_tfidf1_features)

####1gram+2gram
hit_tfidf2 = TfidfVectorizer(sublinear_tf=True, norm='l2', encoding='UTF-8', max_features=1000, ngram_range=(1,2), stop_words='english')
hit_tfidf2_features = hit_tfidf2.fit_transform(hit.Report).toarray()
print(hit_tfidf2_features.shape)
hit_tfidf2_features = pd.DataFrame(hit_tfidf2_features)

####1gram+2gram+3gram
hit_tfidf3 = TfidfVectorizer(sublinear_tf=True, norm='l2', encoding='UTF-8', max_features=1000, ngram_range=(1,3), stop_words='english')
hit_tfidf3_features = hit_tfidf3.fit_transform(hit.Report).toarray()
print(hit_tfidf3_features.shape)
hit_tfidf3_features = pd.DataFrame(hit_tfidf3_features)
```

```python
#Save artifacts to file

##Pre-Processed full dataset
with open('data.pkl', 'wb') as filehandle:
    pickle.dump(data, filehandle)

##Pre-Processed pitchers full dataset
with open('pit_data.pkl', 'wb') as filehandle:
    pickle.dump(pit, filehandle)
    
##Pre-Processed hitters full dataset
with open('hit_data.pkl', 'wb') as filehandle:
    pickle.dump(hit, filehandle)
    
##Pitcher tokenized text
with open('pit_tokens.pkl', 'wb') as filehandle:
    pickle.dump(pit_tokens, filehandle)
    
##Hitter tokenized text
with open('hit_tokens.pkl', 'wb') as filehandle:
    pickle.dump(hit_tokens, filehandle)
    
##Various tfidf vectorizers
with open('pit_tfidf1.pkl', 'wb') as filehandle:
    pickle.dump(pit_tfidf1, filehandle)
with open('pit_tfidf2.pkl', 'wb') as filehandle:
    pickle.dump(pit_tfidf2, filehandle)
with open('pit_tfidf3.pkl', 'wb') as filehandle:
    pickle.dump(pit_tfidf3, filehandle)
with open('hit_tfidf1.pkl', 'wb') as filehandle:
    pickle.dump(hit_tfidf1, filehandle)
with open('hit_tfidf2.pkl', 'wb') as filehandle:
    pickle.dump(hit_tfidf2, filehandle)
with open('hit_tfidf3.pkl', 'wb') as filehandle:
    pickle.dump(hit_tfidf3, filehandle)
    
##Various tfidf vectorized datasets
with open('pit_tfidf1_features.pkl', 'wb') as filehandle:
    pickle.dump(pit_tfidf1_features, filehandle)
with open('pit_tfidf2_features.pkl', 'wb') as filehandle:
    pickle.dump(pit_tfidf2_features, filehandle)
with open('pit_tfidf3_features.pkl', 'wb') as filehandle:
    pickle.dump(pit_tfidf3_features, filehandle)
with open('hit_tfidf1_features.pkl', 'wb') as filehandle:
    pickle.dump(hit_tfidf1_features, filehandle)
with open('hit_tfidf2_features.pkl', 'wb') as filehandle:
    pickle.dump(hit_tfidf2_features, filehandle)
with open('hit_tfidf3_features.pkl', 'wb') as filehandle:
    pickle.dump(hit_tfidf3_features, filehandle)
```

```python
#Summary Stats

print('The number of pitchers with reports is: ', len(pit))
print('The number of hitters with reports is: ', len(hit))
print('The number of original labels is: ', data['FV'].nunique())
print('The number of revised labels is: ', data['FV_r'].nunique())

pit_tokens_lengths = []
for i in pit_tokens:
    pit_tokens_lengths.append(len(i))   
print('The average word length of each pitcher report is:', np.mean(pit_tokens_lengths))

hit_tokens_lengths = []
for i in hit_tokens:
    hit_tokens_lengths.append(len(i))   
print('The average word length of each hitter report is:', np.mean(hit_tokens_lengths))

#Pitcher distribution of labels

grouped_data = pit.groupby(['FV_r'], as_index=False)['Report'].count()
print(grouped_data)
grouped_data.plot.bar(x='FV_r', y='Report', rot=0)
plt.title('Pitcher Distribution of New Labels')
plt.xlabel('Future Value Category')
plt.ylabel('Count')
plt.savefig('pitcher_label_distribution.png')

#Hitter distribution of labels

grouped_data = hit.groupby(['FV_r'], as_index=False)['Report'].count()
print(grouped_data)
grouped_data.plot.bar(x='FV_r', y='Report', rot=0)
plt.title('Hitter Distribution of New Labels')
plt.xlabel('Future Value Category')
plt.ylabel('Count')
plt.savefig('hitter_label_distribution.png')
```
